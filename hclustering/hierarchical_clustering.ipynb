# Hierarchical Clustering

The implementation can be found on the stox library. 

```python
import itertools

from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
from sklearn.preprocessing import normalize
import pandas as pd
import csv

import stox.hierarchical_clustering.clusterizer as clusterizer

%reload_ext autoreload
%load_ext autoreload


```

```python
def show_clusters(clust, x_index=0, y_index=1):
    colors =['green', 'blue', 'red', 'yellow' ]
    for index, c in enumerate(clust):
        X = c.get_features(x_index)
        Y = c.get_features(y_index)
        plt.scatter(X, Y, s = 100, c=colors[index])
    plt.show()
```

```python
points = []
with open("iris.csv", 'r') as f:
    for index, tokens in enumerate(csv.reader(f)):
        if index > 0:
            points.append(tuple([float(t) for t in tokens[:-1]]))
clust = clusterizer.make_clusters(points, no_clusters=3)
show_clusters(clust, x_index=0, y_index=1)
```

```python
points = []
with open("breast_cancer.csv", 'r') as f:
    for index, tokens in enumerate(csv.reader(f)):
        if index > 0:
            points.append(tuple([float(t) for t in tokens[:-1]]))
        if index > 550: break
points = normalize(points)
clust = clusterizer.make_clusters(points, no_clusters=2)
```

```python
for i1, i2 in itertools.combinations(range(9), 2):
    show_clusters(clust, x_index=i1, y_index=i2)
    
```
